```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("missingvals.csv")
```

```python
# Insert the mean of all values for the ones which are missing. Then, drop the column with missing values and add the new column.

from sklearn.impute import SimpleImputer

imputer = SimpleImputer() #missing_values=np.nan, strategy='mean' as default

print(data)
age = data.iloc[:,2:4].values

imputer = imputer.fit(age)
age = imputer.transform(age)

age = age[:,-1:]

data = data.drop(['age'],axis=1)

data.insert(3,"age",age)
print(data) 
```

```python
# Convert to one hot encoder and add to the dataframe

from sklearn import preprocessing

le = preprocessing.LabelEncoder()

country = data.iloc[:,0:1].values

country[:,0] = le.fit_transform(country)
print(country)

ohe = preprocessing.OneHotEncoder()

country = ohe.fit_transform(country).toarray()
print(country)

country = pd.DataFrame(data=country, index=range(22), columns = ['fr','tr','us'])
print(country)

data = data.drop(['country'],axis=1)
data = pd.concat([country,data],axis=1)

print(data)
```
